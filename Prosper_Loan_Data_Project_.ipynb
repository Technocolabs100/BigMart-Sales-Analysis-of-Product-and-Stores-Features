# Handling missing values

```python
data.isnull().sum()
```

```python
null_columns = data.columns[data.isnull().any()]
len(null_columns)
```

There are 43 columns having null values.

# Percentage of missing values

```python
cols = pd.DataFrame(data.isnull().mean().round(4) * 100, columns =['percentage_missing_value']).sort_values(by=['percentage_missing_value'])
print(cols)
```

Here we can see that , there are bunch of columns having more than 50% of null values. those columns are not significant for our analysis so we can remove those columns

```python
#  Remove columns with more than 50% missing values
threshold = 0.5 * len(data)  # 75% threshold
data.dropna(thresh=threshold, axis=1, inplace=True)
```

Imputing the remaining columns having null values.

```python
# Impute missing values for numerical columns with mean
numerical_cols = data.select_dtypes(include=['number']).columns
data[numerical_cols] = data[numerical_cols].fillna(data[numerical_cols].mean())

# Impute missing values for categorical columns with mode
categorical_cols = data.select_dtypes(exclude=['number']).columns
data[categorical_cols] = data[categorical_cols].fillna(data[categorical_cols].mode().iloc[0])
```

```python
# Verify that all missing values have been imputed
missing_values = data.isnull().sum()
print("Missing Values After Imputation:")
print(missing_values[missing_values > 0])
```

# Converting the loan status column into binary target

```python
data.LoanStatus.value_counts()
```

```python
# Define a mapping dictionary
status_mapping = {
    'Completed': 1,
    'Current': 1,
    'Past Due (1-15 days)': 1,
    'Defaulted': 0,
    'Chargedoff': 0,
    'Past Due (16-30 days)': 1,
    'Cancelled': 0,
    'Past Due (61-90 days)': 1,
    'Past Due (31-60 days)': 1,
    'Past Due (91-120 days)': 1,
    'FinalPaymentInProgress': 1,
    'Past Due (>120 days)': 1
}

# Create a new binary target column based on the mapping
data['Loan_Approval'] = data['LoanStatus'].map(status_mapping)
```

```python
data.Loan_Approval.value_counts()
```

# Encoding

```python
for col in categorical_cols:
  print(col, len(data[col].value_counts()))
```

```python
# one hot encoding
columns_to_OHE= ['LoanStatus','EmploymentStatus','IncomeRange', 'ProsperRating (Alpha)']
data= pd.get_dummies(data, columns=columns_to_OHE)
data.head()
```

```python
# Lable encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
columns_to_encode = ['ListingKey', 'ListingCreationDate', 'BorrowerState', 'Occupation', 'IsBorrowerHomeowner', 'CurrentlyInGroup', 'DateCreditPulled', 'FirstRecordedCreditLine',
                     'IncomeVerifiable', 'LoanKey', 'LoanOriginationDate', 'LoanOriginationQuarter', 'MemberKey']


# Apply label encoding to each specified column and replace the original values
for column in columns_to_encode:
    data[column] = le.fit_transform(data[column])
```
