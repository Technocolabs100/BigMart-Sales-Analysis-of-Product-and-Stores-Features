Loading Packages and Data

Replacing NAN with mean looks like our best option


```python
mean_w = train_df['Item_Weight'].mean()

# Replace NaN values in column 'A' with the mean
train_df['Item_Weight'].fillna(mean_w, inplace=True)


mean_w1 = test_df['Item_Weight'].mean()

# Replace NaN values in column 'A' with the mean
test_df['Item_Weight'].fillna(mean_w1, inplace=True)
```

Dropping Outlet_Identifier

```python
del train_df['Outlet_Identifier']
del test_df['Outlet_Identifier']
```

```python
train_df['Item_Fat_Content'].replace('reg', 'Regular', inplace=True)  
train_df['Item_Fat_Content'].replace('LF', 'Low Fat', inplace=True)  
```

Label encoding and one hot encoding for nominal categorical variables 

```python
label_encoder = LabelEncoder()

# Fit and transform the 'Category' column
train_df['item_fat_encoded'] = label_encoder.fit_transform(train_df['Item_Fat_Content'])
train_df['size_ncoded'] = label_encoder.fit_transform(train_df['Outlet_Size'])
train_df['loctype_encoded'] = label_encoder.fit_transform(train_df['Outlet_Location_Type'])
train_df['outlet_type_encoded'] = label_encoder.fit_transform(train_df['Outlet_Type'])
train_df.drop(['Item_Fat_Content', 'Outlet_Size','Outlet_Type','Outlet_Location_Type'], axis=1, inplace=True)
```

```python
train_df=pd.get_dummies(train_df, columns=['Item_Type'])
```

```python
train_df
```

```python

# train_df.drop(['Item_Weight'], axis=1, inplace=True)
```

```python
train_df.drop(['Outlet_Establishment_Year'], axis=1, inplace=True)
```

```python
train_df
```

```python
scaler = StandardScaler()

# Specify the columns you want to standardize
columns_to_standardize = ['Item_Visibility', 'Item_MRP']

# Fit and transform the specified columns
train_df[columns_to_standardize] = scaler.fit_transform(train_df[columns_to_standardize])
```

Splitting 

```python
y = train_df['Item_Outlet_Sales']
X = train_df.drop('Item_Outlet_Sales', axis = 1)

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.8, random_state = 0)
```

# LINEAR REGRESSION


```python
# Create a LinearRegression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_val)
```

```python

# Calculate the mean squared error
mse = mean_squared_error(y_val, y_pred)
print("Mean Squared Error:", mse)
```

```python
r_squared = r2_score(y_val, y_pred)

print("R-squared:", r_squared)
```

# RIDGE AND LASSO REGREESION

```python
# You can specify the regularization strength (alpha) here. Higher alpha values will lead to stronger regularization.
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
y_pred = ridge.predict(X_val)
mse = mean_squared_error(y_val, y_pred)
print(f"Mean Squared Error: {mse}")
coefficients = ridge.coef_
print("Coefficients:", coefficients)
r_squared = r2_score(y_val, y_pred)

print("R-squared:", r_squared)
```

```python
lasso = Lasso(alpha=1.0)

# Fit the model to the training data
lasso.fit(X_train, y_train)

# Make predictions on the test data
y_pred = lasso.predict(X_val)

# Evaluate the model using Mean Squared Error
mse = mean_squared_error(y_val, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
r_squared = r2_score(y_val, y_pred)

print("R-squared:", r_squared)
```

# RF

```python
reg = RandomForestRegressor(n_estimators=100, random_state=42)
reg.fit(X_train, y_train)  # For regression tasks
y_pred = reg.predict(X_val)  # For regression tasks
mse = mean_squared_error(y_val, y_pred)
print(f"Mean Squared Error: {mse}")
r_squared = r2_score(y_val, y_pred)

print("R-squared:", r_squared)
```

# XGB

```python
reg = XGBRegressor(n_estimators=100, random_state=42)
reg.fit(X_train, y_train)  # For regression tasks
y_pred = reg.predict(X_val)  # For regression tasks
mse = mean_squared_error(y_val, y_pred)
print(f"Mean Squared Error: {mse}")
r_squared = r2_score(y_val, y_pred)

print("R-squared:", r_squared)
```
